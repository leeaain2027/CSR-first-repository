```python
# 필요한 패키지 import 하기
import os 
import cv2 
import matplotlib.pyplot as plt
import numpy as np
import dlib
```

```python
# 사진 불러오기 + 화면에 보여주기
home_dir = os.getenv('HOME')
my_image_path = os.path.join(home_dir, 'work/camera_sticker/images/input/image1.jpg')
img_bgr = cv2.imread(my_image_path)

img_original = img_bgr.copy() # 확인용 실행할 때마다 스티커 연달아서 붙이지 않도록 원본 미리 저장
img_show = img_bgr.copy()
plt.imshow(img_bgr)
plt.show()
```

```python
# RGB로 색 순서 바꿔서 제대로 보여주기
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
# 정면 얼굴 찾는 도구 준비
detector_hog = dlib.get_frontal_face_detector()
dlib_rects = detector_hog(img_rgb, 1) # 1: 한번 확대해서 검사. 더 꼼꼼히 찾기

# 찾은 얼굴 영역 박스 출력
print(dlib_rects)

# 찾은 얼굴 영역에 사각형 그리기
for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top() 
    r = dlib_rect.right() 
    b = dlib_rect.bottom() 

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)  # 얼굴 위치에 초록색 사각형 그리기 (두께 2)

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
# 얼굴의 주요 점(랜드마크)을 찾는 모델 불러오기
model_path = os.path.join(home_dir, 'work/camera_sticker/models/shape_predictor_68_face_landmarks.dat')

landmark_predictor = dlib.shape_predictor(model_path) # 얼굴 점(랜드마크)을 찾아주는 예측기 
```

```python
list_landmarks = [] # 68개 점 좌표를 저장할 리스트

# face landmark 좌표 만들기
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    # points.parts()로 점들을 꺼내서 (x, y) 형태로 바꿔 저장
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)

print(len(list_landmarks[0])) # 점 개수 확인
```

```python
# 찍힌 점들을 화면에 표시하기 (확인용)
for landmark in list_landmarks:
    for point in landmark:
        # 노란 점
        cv2.circle(img_show, point, 2, (0, 255, 255), -1)

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
# 코 위치(기준점)와 얼굴 크기 정보 뽑기
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print (landmark[30]) # 코 가운데 점 위치
    # 스티커 붙일 좌표. 
    x = landmark[30][0]  
    y = landmark[30][1]
    w = h = dlib_rect.width() # 얼굴 박스 너비를 기준으로 스티커 크기를 정
    print (f'(x,y) : ({x},{y})')
    print (f'(w,h) : ({w},{h})')
```

<br>
▶ 이전 실습에서는 왕관의 위치가 머리 위에 있어야 해서, landmark[30][1] - dlib_rect.height()//2 이렇게 했지만,<br>
이번 과제에서는 수염의 위치가 코쪽에 있어야 하기 때문에, x, y를 코 끝점 위치로 잡았다.<br>
<br>
코 끝점은 얼굴을 인식한 박스를 기준으로 좌표가 찍히는 게 아니라, 이미지 크기 전체를 기준으로 좌표가 찍힌다.<br>
인식 박스는 전체 이미지에서 얼굴을 찾기 위함이고, 얼굴이 여기 있다는 걸 알면 눈, 코, 입이 어디있는지 인식할 수 있기 때문에 박스가 존재하는 것.<br>
<br>
코 좌표 (x, y)는 이미지 전체 기준의 절대 좌표이지만,<br>
그 좌표가 생성되는 기준은 매 이미지마다 인식된 얼굴 위치에 의해 달라진다.<br>
<br>
for문으로 여러 이미지를 처리할 때 각 이미지의 코 좌표는 모두 다르지만,<br>
얼굴 박스를 기준으로 한 상대적 구조(얼굴 안에서의 위치)는 유지되기 때문에<br>
수염은 그 상대적 기준에 의해 안정적으로 붙을 수 있다.<br>
<br>

```python
# 매번 실행할 때 스티커 없는 초기 상태로 되돌리기
img_bgr = img_original.copy()
img_show = img_bgr.copy()

# 스티커 이미지 불러오기 (투명 배경 포함해서)
sticker_path = os.path.join(home_dir, 'work/camera_sticker/images/cat-whiskers.png')
# 투명 배경까지 같이 읽기 위해 IMREAD_UNCHANGED 사용
img_sticker = cv2.imread(sticker_path, cv2.IMREAD_UNCHANGED)

w = int(dlib_rect.width() * 0.8) # 스티커 크기를 얼굴 가로 사이즈의 0.8로 맞춤
h = w

# 스티커 크기 변경
img_sticker = cv2.resize(img_sticker, (w,h))
print (img_sticker.shape)
```

<br>
▶ 스티커 크기를 얼굴 비율에 맞추고 싶어서 0.8로 설정함
<br><br>
▶ 그리고 이전 코드 그대로 사용해서 고양이 수염 붙이려고 하니, 고양이 수염의 배경이 흰색 + 수염은 투명색으로 나왔었다.
색상 변환이 필요해서 cv2.IMREAD_UNCHANGED를 추가함. 이후 여러 코드를 추가할 예정
<br><br>
▶ 확인해본다고 추가 실행할 때마다 스티커가 붙는 번거로움이 있어서 초기 상태로 되돌리는 코드 넣음
<br><br>

```python
# 스티커 안에서 "코가 들어갈 위치"를 대략 지정
sticker_x = int(w * 0.5)
sticker_y = int(h * 0.45)

# 스티커의 왼쪽 위 좌표를 계산 (코 좌표 - 스티커 기준점)
refined_x = x - sticker_x
refined_y = y - sticker_y
print (f'(x,y) : ({refined_x},{refined_y})')
```

<br>
▶ 근데 왜 0.5, 0.45를 했을까? <br>
얼굴의 너비를 봤을 때 정중앙은 코다. 코 지점을 정중앙으로 만들어야 하기 때문에 0.5<br>
얼굴의 세로 길이로는 중앙에서 살짝 위에 있는 게 예쁠 것 같아서 0.45로 설정함<br>
<br>
sticker_x = int(767 × 0.5)  = 383<br>
sticker_y = int(767 × 0.45) = 345<br>
<br>
refined_x = 1095 - 383 = 712<br>
refined_y = 1273 - 345 = 928<br>
<br>
스티커의 왼쪽 위를 (712, 928)에 두면 스티커 안의 (383, 345) 지점이 정확히 (1095, 1273), 즉 코 위치에 오게 된다<br>
<br><br>

```python
# 스티커가 사진 바깥으로 나가는 경우 잘라내기
# 왼쪽 밖으로 나가면 스티커의 왼쪽 부분을 잘라내고 x를 0으로 맞춤
if refined_x < 0:
    img_sticker = img_sticker[:, -refined_x:]
    refined_x = 0
# 위쪽 밖으로 나가면 스티커의 위쪽 부분을 잘라내고 y를 0으로 맞춤
if refined_y < 0:
    img_sticker = img_sticker[-refined_y:, :]
    refined_y = 0

print (f'(x,y) : ({refined_x},{refined_y})')
```

```python
# (위에서 img_show에 합성한 걸) 원본(img_bgr)에도 동일하게 적용. 최종 결과용

# 스티커 높이/너비 다시 가져오기
h, w = img_sticker.shape[:2]
# 원본 이미지에서 붙일 영역 가져오기
sticker_area = img_bgr[refined_y:refined_y+h, refined_x:refined_x+w].astype(np.float32)
# 스티커에서 색 부분(3채널: 파란/초록/빨강)만 꺼내기
sticker_bgr = img_sticker[:, :, :3].astype(np.float32)
# (H,W,1) 스티커에서 투명도를 꺼내 0~1 사이로 만들기. 계산하기 편하게 채널을 하나 늘려서 (세로, 가로, 1)로 만들기
alpha = (img_sticker[:, :, 3].astype(np.float32) / 255.0)[:, :, None]  # (h,w,1)
# 색상 합성 공식
blended = alpha * sticker_bgr + (1 - alpha) * sticker_area
# 원본 이미지에 적용
img_bgr[refined_y:refined_y+h, refined_x:refined_x+w] = blended.astype(np.uint8)

plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```
