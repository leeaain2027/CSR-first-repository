## 키오스크 만들기 (2)

```python
class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i+1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드
    def menu_select(self):
        self.order_menu = []  # 주문 리스트
        self.order_price = []  # 가격 리스트

        r = 1
        while r != 0:
            # 메뉴 선택
            n = 0  # 기본값 넣음
            while n < 1 or len(self.menu) < n:
                n = int(input("음료 번호를 입력해주세요 : \n >>>"))

                if 1 <= n & n <= len(self.menu):
                    print(self.menu[n-1], ' : ', self.price[n-1], '원')  # 메뉴 출력

                else:  
                    print("없는 메뉴입니다. 다시 주문해주세요.")

            # HOT/ICE 선택
            t = 0  # 기본값 넣음
            while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
                t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력해주세요 : \n >>>"))
                        
                if t == 1:
                    self.temp = 'HOT'
                    print(self.temp)
                
                elif t == 2:
                    self.temp = 'ICE'
                    print(self.temp)

                else:    
                    print("1과 2 중 하나를 입력해주세요.")

            self.order_menu.append(self.temp + ' ' + self.menu[n-1])  # 추가 메뉴를 리스트에 추가
            self.order_price.append(self.price[n-1])  # 추가 메뉴 가격 리스트에 추가
            
            # 추가 주문할지 물어봄
            r = int(input("추가 주문 없이 결제를 원하시면 0을 입력해주세요. : \n >>>"))

            if r == 0:
                print(self.order_menu, self.order_price)

            
    # 지불
    def pay(self):
        p = 0
        while p != 1 and p != 2: 
            p = int(input('지불 수단을 입력해주세요. 현금이면 1, 카드면 2 : \n >>>'))

            if p == 1:
                print('직원을 호출하겠습니다.')
            elif p == 2:
                print('IC칩 방향에 맞게 카드를 꽂아주세요.')
            else:
                print('다시 결제를 시도해 주세요.')

    # 주문서 출력 
    def table(self):
        menu_temp = self.order_menu 
        price_temp = self.order_price 

        # 외곽
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        # 주문메뉴 : 금액
        for i in range(len(menu_temp)):
            print(menu_temp[i], ' : ', price_temp[i])

        print('합계 금액 :', sum(price_temp))

        # 외곽
        for i in range(5):
            print('|' + ' ' * 31+ '|')
        print('⟝' + '-' * 30 + '⟞')



a = Kiosk()  # 객체 생성 
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```
