## 사각형 넓이 구하기

```python
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            result_r = self.rect()
            print(f"직사각형 함수는 rect()입니다.\n결과: {result_r}")

        elif self.square == 2:
            result_p = self.par()
            print(f'평행사변형 함수는 par()입니다.\n결과: {result_p}')
                    
        elif self.square == 3:
            result_t = self.trape()
            print(f'사다리꼴 함수는 trape()입니다.\n결과: {result_t}')

        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 ' + str(area)
        return result

    def par(self):
        base_line, height = map(int, input('밑변, 높이를 입력하세요. 예시 : 밑변,높이\n >>>').split(','))
        area = base_line * height
        result = '평행사변형의 넓이는 ' + str(area)
        return result

    def trape(self):
        base_line, top_line, height = map(int, input('밑변, 윗변, 높이를 입력하세요. 예시 : 밑변,윗변,높이\n >>>').split(','))
        area = (base_line + top_line) * height * 0.5
        result = '사다리꼴 넓이는 ' + str(area)
        return result

a = Square()
```
