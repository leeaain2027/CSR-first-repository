# quest_1 문제 구현

```python
import random as r

class Account:
    count = 0
    count_d = 0

    # 계좌 정보 초기 설정
    def __init__(self, name, balance):
        self.bank = "SC은행"
        self.name = name
        self.number = str(r.randint(0,1000)).zfill(3)+"-"+str(r.randint(0,100)).zfill(3)+"-"+str(r.randint(0,1000000)).zfill(3)
        self.balance = balance

        Account.count += 1  # 객체 생성할 때마다 개수 +1
        self.get_account_num()
    
    def get_account_num(self):
        print(f"계좌의 개수는 {Account.count}입니다.")
        
    # 입금 메서드
    def deposit(self, d):
        Account.count_d += 1
        
        if d <= 0:
            raise ValueError("입금은 최소 1원 이상이어야 합니다.")
        else:
            self.balance += d

        # 이자 지급하기
        if Account.count_d == 5:
            self.balance += self.balance * 0.01

    # 출금 메서드
    def withdraw(self, w):
        if w > self.balance:
            raise ValueError("계좌 잔고 이상의 금액을 출금할 수 없습니다.")
        else:
            self.balance -= w

    # 정보 출력 메서드
    def display_info(self):
        print(f"은행이름: {self.bank}, 예금주: {self.name}, 계좌번호: {self.number}, 잔고: {self.balance:,}")
```

```python
# 인스턴스
client_1 = Account("파이썬", 10000)
client_2 = Account("최사랑", 10000000)
client_3 = Account("곰", 1)

# 입금, 출금 확인
client_1.deposit(5000)
client_1.deposit(5000)
client_1.deposit(5000)
client_1.deposit(5000)
client_1.deposit(1)
client_1.withdraw(10000)
print(client_1.balance)
```

```python
# 인스턴스 중에 100만원 이상인 고객 정보 출력
for x in [client_1, client_2, client_3]:
    if x.balance >= 1000000:
        x.display_info()
```

# chatGPT quest_1 문제 풀이

```python
import random


class Account:
    # Q2. 클래스 변수 (계좌 개수)
    account_count = 0

    def __init__(self, owner, balance):
        # Q1. 기본 정보
        self.bank = "SC은행"
        self.owner = owner
        self.balance = balance

        # 계좌번호 랜덤 생성 (3-2-6)
        first = random.randint(100, 999)
        second = random.randint(10, 99)
        third = random.randint(100000, 999999)
        self.account_number = f"{first}-{second}-{third}"

        # 입금 횟수
        self.deposit_count = 0

        # Q10. 입출금 내역
        self.deposit_log = []
        self.withdraw_log = []

        # 계좌 개수 증가
        Account.account_count += 1

    # Q3. 계좌 개수 출력
    @classmethod
    def get_account_num(cls):
        return cls.account_count

    # Q4. 입금
    def deposit(self, amount):
        if amount < 1:
            print("입금은 1원 이상만 가능합니다.")
            return

        self.balance += amount
        self.deposit_count += 1
        self.deposit_log.append(amount)

        # Q7. 5회 입금마다 이자 지급
        if self.deposit_count % 5 == 0:
            interest = self.balance * 0.01
            self.balance += interest

    # Q5. 출금
    def withdraw(self, amount):
        if amount > self.balance:
            print("잔액이 부족합니다.")
            return

        self.balance -= amount
        self.withdraw_log.append(amount)

    # Q6. 정보 출력
    def display_info(self):
        print(f"은행이름: {self.bank}")
        print(f"예금주: {self.owner}")
        print(f"계좌번호: {self.account_number}")
        print(f"잔고: {self.balance:,.0f}원")

    # Q10. 입금 내역 출력
    def deposit_history(self):
        print("입금 내역:", self.deposit_log)

    # Q10. 출금 내역 출력
    def withdraw_history(self):
        print("출금 내역:", self.withdraw_log)
```

```python
# Q8. 여러 객체 생성 & 리스트 저장
accounts = [
    Account("파이썬", 1_500_000),
    Account("자바", 800_000),
    Account("자바스크립트", 2_000_000)
]
```

```python
# Q9. 잔고 100만원 이상 고객만 출력
for acc in accounts:
    if acc.balance >= 1_000_000:
        acc.display_info()
        print("-" * 30)
```

```python
# 사용 예시 (입출금 & 이자 확인)
a = Account("테스트", 10000)

for _ in range(5):
    a.deposit(1000)   # 5번째 입금 시 이자 발생

a.withdraw(3000)

a.display_info()
a.deposit_history()
a.withdraw_history()

print("총 계좌 수:", Account.get_account_num())
```

▶ 이건 대홍님이 알려주신 좋은 팁.

그래서 randint 쓰고 zfill 쓰면 000, 001, 010,.. 가능!
