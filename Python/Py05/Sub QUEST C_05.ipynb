```python
import random as r

class Account:
    count = 0
    count_d = 0

    # 계좌 정보 초기 설정
    def __init__(self, name, balance):
        self.bank = "SC은행"
        self.name = name
        self.number = str(r.randint(111,1000))+"-"+str(r.randint(11,100))+"-"+str(r.randint(11,1000000))
        self.balance = balance

        Account.count += 1  # 객체 생성할 때마다 개수 +1
        self.get_account_num()
    
    def get_account_num(self):
        print(f"계좌의 개수는 {Account.count}입니다.")
        
    # 입금 메서드
    def deposit(self, d):
        Account.count_d += 1
        
        if d <= 0:
            raise ValueError("입금은 최소 1원 이상이어야 합니다.")
        else:
            self.balance += d

        # 이자 지급하기
        if Account.count_d == 5:
            self.balance += self.balance * 0.01

    # 출금 메서드
    def withdraw(self, w):
    #     조건:
    #         계좌 잔고 이상으로 출금 x
        if w > self.balance:
            raise ValueError("계좌 잔고 이상의 금액을 출금할 수 없습니다.")
        else:
            self.balance -= w

    # 정보 출력 메서드
    def display_info(self):
        print(f"은행이름: {self.bank}, 예금주: {self.name}, 계좌번호: {self.number}, 잔고: {self.balance:,}")
```

```python
# 인스턴스
client_1 = Account("파이썬", 10000)
client_2 = Account("최사랑", 10000000)
client_3 = Account("곰", 1)

# 입금, 출금 확인
client_1.deposit(5000)
client_1.deposit(5000)
client_1.deposit(5000)
client_1.deposit(5000)
client_1.deposit(1)
client_1.withdraw(10000)
print(client_1.balance)
```

```python
# 인스턴스 중에 100만원 이상인 고객 정보 출력
for x in [client_1, client_2, client_3]:
    if x.balance >= 1000000:
        x.display_info()
```
