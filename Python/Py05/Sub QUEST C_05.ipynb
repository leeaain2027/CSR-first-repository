```python
import random


class Account:
    account_count = 0

    def __init__(self, owner, balance):
        self.bank = "SCì€í–‰"
        self.owner = owner
        self.balance = balance

        # ê³„ì¢Œë²ˆí˜¸ ìƒì„±
        self.account_number = f"{random.randint(100,999)}-{random.randint(10,99)}-{random.randint(100000,999999)}"

        # ê¸°ë¡ ê´€ë ¨
        self.deposit_count = 0
        self.deposit_log = []
        self.withdraw_log = []
        self.transfer_log = []  # ğŸ”¥ ì´ì²´ ì´ë ¥

        Account.account_count += 1

    @classmethod
    def get_account_num(cls):
        return cls.account_count

    def deposit(self, amount):
        if amount < 1:
            print("ì…ê¸ˆì€ 1ì› ì´ìƒë§Œ ê°€ëŠ¥í•©ë‹ˆë‹¤.")
            return

        self.balance += amount
        self.deposit_count += 1
        self.deposit_log.append(amount)

        # 5íšŒ ì…ê¸ˆë§ˆë‹¤ ì´ì
        if self.deposit_count % 5 == 0:
            interest = self.balance * 0.01
            self.balance += interest

    def withdraw(self, amount):
        if amount > self.balance:
            print("ì”ì•¡ ë¶€ì¡±ìœ¼ë¡œ ì¶œê¸ˆ ì‹¤íŒ¨")
            return False

        self.balance -= amount
        self.withdraw_log.append(amount)
        return True

        # ğŸ”¥ ì´ì²´ ê¸°ëŠ¥
    def transfer(self, target_account, amount):
        if amount <= 0:
            print("ì´ì²´ ê¸ˆì•¡ì€ 1ì› ì´ìƒì´ì–´ì•¼ í•©ë‹ˆë‹¤.")
            return
    
        # ğŸ”¹ ì´ì²´ ì „ ì”ê³  ì €ì¥
        sender_before = self.balance
        receiver_before = target_account.balance
    
        if self.withdraw(amount):
            target_account.deposit(amount)
    
            sender_after = self.balance
            receiver_after = target_account.balance
    
            # ğŸ”¹ ë³´ë‚´ëŠ” ì‚¬ëŒ ë¡œê·¸
            self.transfer_log.append(
                f"""
    [ì´ì²´ ì¶œê¸ˆ]
    ì´ì²´ ì „ ì”ê³  : {sender_before:,.0f}ì›
    â†’ {target_account.owner}({target_account.account_number}) {amount:,.0f}ì› ì´ì²´
    ì´ì²´ í›„ ì”ê³  : {sender_after:,.0f}ì›
    """.strip()
            )
    
            # ğŸ”¹ ë°›ëŠ” ì‚¬ëŒ ë¡œê·¸
            target_account.transfer_log.append(
                f"""
    [ì´ì²´ ì…ê¸ˆ]
    ì´ì²´ ì „ ì”ê³  : {receiver_before:,.0f}ì›
    â† {self.owner}({self.account_number}) {amount:,.0f}ì› ì…ê¸ˆ
    ì´ì²´ í›„ ì”ê³  : {receiver_after:,.0f}ì›
    """.strip()
            )
        else:
            print("ì´ì²´ ì‹¤íŒ¨ (ì”ì•¡ ë¶€ì¡±)")
    
    def display_info(self):
        print(f"ì€í–‰: {self.bank}")
        print(f"ì˜ˆê¸ˆì£¼: {self.owner}")
        print(f"ê³„ì¢Œë²ˆí˜¸: {self.account_number}")
        print(f"ì”ê³ : {self.balance:,.0f}ì›")

    # ğŸ” ì´ì²´ ë‚´ì—­ í™•ì¸
    def transfer_history(self):
        print(f"\n[{self.owner} ì´ì²´ ë‚´ì—­]")
        if not self.transfer_log:
            print("ì´ì²´ ë‚´ì—­ ì—†ìŒ")
        else:
            for log in self.transfer_log:
                print(log)
                print("-" * 30)
```
