# 사용자로부터 두 개의 정수와 연산자(+,-,*,/,**)를 입력받아 계산 결과를 출력하는 프로그램을 만들어주세요

**조건**

- 사용자가 입력한 값이 정수가 아닌 경우 ValueError를 처리하여 적절한 오류 메시지를 출력해야 합니다.

- 정수가 입력될 때 까지 “잘못된 입력입니다. 정수를 입력해주세요.”를 출력하며 다시 입력받기를 시도합니다.

- 나눗셈 연산 시 두 번째 정수가 0인 경우 ZeroDivisionError를 처리하여 적절한 오류 메시지를 출력합니다.

- 사용자가 지원하지 않는 연산자를 입력한 경우 적절한 오류 메시지를 출력합니다.

- math 모듈을 사용하여 제곱 연산(**)을 지원합니다.

- 계산기를 통해 계속 계산을 할 것인지 입력받습니다.

```python
import math

def input_integer(prompt):
    """정수를 입력 받고 유효성 검사를 수행함"""
    while True:
        try:
            value = int(input(prompt))
            return value
        except ValueError:
            print("잘못된 입력입니다. 정수를 입력해 주세요.")

def input_operator():
    """연산자를 입력받음"""
    return input("연산자를 입력하세요 (+, -, *, /, **): ")

def calculator(a, b, operator):
    """계산 수행"""
    if operator == "+":
        return a + b
    elif operator == "-":
        return a - b
    elif operator == "*":
        return a * b
    elif operator == "/":
        if b == 0:
            raise ZeroDivisionError
        return a / b
    elif operator == "**":
        return math.pow(a, b)
    else:
        raise ValueError("지원하지 않는 연산자입니다.")

def ask_continue():
    """계속 계산할지 여부 확인"""
    answer = input("계속 계산하시겠습니까? (y/n): ").lower()
    return answer == "y"

def main():
    print("=== 계산기 프로그램 ===")
    while True:
        a = input_integer("첫 번째 정수를 입력하세요: ")
        b = input_integer("두 번째 정수를 입력하세요: ")
        operator = input_operator()

        try:
            result = calculator(a, b, operator)
            print("결과:", result)
        except ZeroDivisionError:
            print("오류: 0으로 나눌 수 없습니다.")
        except ValueError as e:
            print("오류:", e)

        if not ask_continue():
            print("프로그램을 종료합니다.")
            break

main()
```
